```python
import time
import numpy as np
from skimage import io
import plotly.graph_objects as go
import os
```

```python
import os
from PIL import Image

# 输入文件夹和输出文件夹路径
input_folder = r"D:\coding\AcousticSensing\Data\NewCylinder\CTpic\X光CT\三个方向切片\上切片"  # 替换为包含彩色图像的文件夹路径
output_folder = r"D:\coding\AcousticSensing\Data\NewCylinder\CTpic\ctpicture_up"  # 替换为保存灰度图像的文件夹路径

# 确保输出文件夹存在
os.makedirs(output_folder, exist_ok=True)

# 遍历输入文件夹中的所有文件
for filename in os.listdir(input_folder):
    if filename.endswith(".tif"):  # 仅处理.tif图像文件
        # 打开.tif图像
        tif_image = Image.open(os.path.join(input_folder, filename))
        
        # 将.tif图像转换为灰度图像
        gray_image = tif_image.convert("L")
        
        # 保存灰度图像到输出文件夹
        output_path = os.path.join(output_folder, filename)
        gray_image.save(output_path)

print("处理完成。")
```

```python
path=r"D:\coding\AcousticSensing\Data\NewCylinder\CTpic\ctpicture_up"
vol=tif_collection = io.imread(os.path.join(path, '*.tif'))
volume=vol.T
r,c=volume[0].shape
print(volume.shape)
frames=375
```

```python
fig = go.Figure(frames=[go.Frame(data=go.Surface(
    z=(37.4 - k * 0.1) * np.ones((r, c)),
    surfacecolor=np.flipud(volume[frames - 1 - k]),
    cmin=0, cmax=200
    ),
    name=str(k) # you need to name the frame for the animation to behave properly
    )
    for k in range(frames)
    ]
)

fig.add_trace(go.Surface(
    z=37.4 * np.ones((r, c)),
    surfacecolor=np.flipud(volume[frames-1]),
    colorscale='Gray',
    cmin=0, cmax=200,
    colorbar=dict(thickness=20, ticklen=4)
))

def frame_args(duration):
    return {
            "frame": {"duration": duration},
            "mode": "immediate",
            "fromcurrent": True,
            "transition": {"duration": duration, "easing": "linear"},
        }

sliders = [
            {
                "pad": {"b": 10, "t": 60},
                "len": 0.9,
                "x": 0.1,
                "y": 0,
                "steps": [
                    {
                        "args": [[f.name], frame_args(0)],
                        "label": str(k),
                        "method": "animate",
                    }
                    for k, f in enumerate(fig.frames)
                ],
            }
        ]

# Layout
fig.update_layout(
         title='Slices in volumetric data',
         width=600,
         height=600,
         scene=dict(
                    zaxis=dict(range=[-0.1, 37.5], autorange=False),
                    aspectratio=dict(x=1, y=1, z=1),
                    ),
         updatemenus = [
            {
                "buttons": [
                    {
                        "args": [None, frame_args(50)],
                        "label": "&#9654;", # play symbol
                        "method": "animate",
                    },
                    {
                        "args": [[None], frame_args(0)],
                        "label": "&#9724;", # pause symbol
                        "method": "animate",
                    },
                ],
                "direction": "left",
                "pad": {"r": 10, "t": 70},
                "type": "buttons",
                "x": 0.1,
                "y": 0,
            }
         ],
         sliders=sliders
)

fig.show()
```
